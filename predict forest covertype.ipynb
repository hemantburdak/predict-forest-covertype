```python
def class_likelihood_params(class_training_data):
    
    class_likelihood_params_dict = dict()

    for cols in class_training_data.columns[0:8]:

        empty_dict = dict()

        if class_training_data[cols].dtype == "int64":

            for value in data[cols].unique():

                if value not in class_training_data[cols].unique():
          
                  empty_dict[value] = 1/(len(class_training_data)+len(data[cols].unique()))

                else:
          
                  empty_dict[value] = len(class_training_data[class_training_data[cols] == value])/len(class_training_data)
        


        else:

            empty_dict["class_normal_mean"] = np.mean(class_training_data[cols])
            empty_dict["class_normal_std"] = np.std(class_training_data[cols]) 

        class_likelihood_params_dict[cols] = empty_dict

    return class_likelihood_params_dict
```

```python
from concurrent.futures import ThreadPoolExecutor
```

```python
with ThreadPoolExecutor(max_workers=8) as p:
    
    likelihood_params_dicts = list()
    
    lazy_likelihood_params_dicts = [p.submit(class_likelihood_params,training_data[training_data["feature54"] == x]) for x in range(1,8)]
    
    for params_dict in lazy_likelihood_params_dicts:
        
        likelihood_params_dicts.append(params_dict.result())
```

```python
likelihood_params_dicts
```

```python
import scipy.stats as s
```

```python
def class_conditioned_likelihood_per_row(params_dict, cv_data_row):
    
    p_features_on_class = 1

    for cols in cv_data.columns[0:8]:

        if cv_data[cols].dtype == "int64":

            p_features_on_class = p_features_on_class * params_dict[cols][cv_data_row[cols]]

        else:

            p_features_on_class = p_features_on_class * s.norm.pdf(cv_data_row[cols],
                                                               params_dict[cols]["class_normal_mean"],
                                                               params_dict[cols]["class_normal_std"])
            
    return p_features_on_class
```

```python
priors = list()

for i in range (1,8):

    priors.append(len(training_data[training_data['feature54'] == i])/len(training_data))
```

```python
priors
```

```python
def fetch_predictions(data):
    
    predictions = list()
    
    for i in range(0,data.shape[0]):
        
        posterior_i = list()
        
        for j in range(0,7):
            
            posterior_i.append(class_conditioned_likelihood_per_row(likelihood_params_dicts[j],data.iloc[i,0:8])*priors[j])
            
        posterior_i = np.array(posterior_i)
        
        predictions.append(np.argmax(posterior_i))
        
    return np.array(predictions)
```
